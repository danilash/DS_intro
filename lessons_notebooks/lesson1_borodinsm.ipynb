## Мини-проект 1

```python
x=int(input("Введи число 1="))
y=int(input("Введи число 2="))
z=str(input("Введи действия с числами (сложение/вычитание/умножение/деление/степень) без регистра="))
if z=="сложение":
  print(f"{x}+{y}={x+y}")
elif z=="вычитание":
  print(f"{x}-{y}={x-y}")
elif z=="умножение":
  print(f"{x}*{y}={x*y}")
elif z=="деление":
  print(f"{x}/{y}={x/y}")
else:
  print(f"{x}**{y}={x**y}")
```

```python
# ax^2+bx+c=0
a=int(input("Введи a="))
b=int(input("Введи b="))
c=int(input("Введи c="))

d=b**2-4*a*c
if d>0:
  x1=(-b+d**0.5)/(2*a)
  x2=(-b-d**0.5)/(2*a)
  print("1 корень = ",x1)
  print("2 корень = ",x2)
elif d==0:
  x1=-b/2*a
  print("Корень =",x1)
else:
  print("Нет корней")
```

```python
## 3 Задание

print("Поиск подъезда и этажа по номеру квартиры")
print("(Примечание! В доме всего 5 подъездов по 20 квартир в каждой, если введёте номер квартиры >100 или <=0, то выдаст ошибку)")

x=int(input("Введи номер квартиры="))
if x>100 or x<=0:    ## Тест на дурака
  print("Ошибка!")
else:
  for i in range(1,6):
    if x>(i-1)*20 and x<=i*20:
      p=i
  n=(p-1)*20
  e=0
  while n!=x+1:
    if n%4==0 and n!=x:
      e+=1
    n+=1
  print(f"Квартира №{x} находится в подъезде №{p}, на этаже №{e}")

```
